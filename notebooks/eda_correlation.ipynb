```python
import numpy as np
import xarray as xr
import pandas as pd
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_parquet("../dataset/combined/undersampled/train-sel-10.parquet")
df
```

```python
df_corr = df.drop(columns=["zone_id"]).corr()
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))

plt.figure(figsize=(10, 6))
sns.heatmap(df_corr, annot=True, cmap="vlag", mask=mask)
```
