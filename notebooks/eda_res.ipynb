## Merge

```python
temp_files = glob.glob(f"../temp/*.parquet")
temp_files
```

```python
dd_final = pd.read_parquet(temp_files[0])
dd_final
```

```python
for f in temp_files[1:]:
    # read next dataset
    dd_temp = pd.read_parquet(f)

    # merge on ZONE_ID
    dd_final = dd_final.merge(dd_temp, on=["zone_id", "ts"]).drop(columns=["target_x"], errors="ignore")
```

```python
dd_final.head()
```

```python
dd_final.columns
```

```python
cols = list(x for x in dd_final.columns if not x.startswith("target_"))
cols.remove("ts")
cols.remove("target")
cols.insert(1, "ts")
cols.append("target")
dd_final.drop(columns=[x for x in dd_final.columns if x.startswith("target_")])
dd_final = dd_final[cols]
```

```python
dd_final.head()
```

```python
dd_final.info()
```

```python
dd_final.to_parquet("meong.parquet", engine="pyarrow")
```
